```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
% matplotlib inline
```

```python
np.random.seed(7)
plt.rcParams['figure.figsize'] = 8, 5
```

```python
data1 = pd.DataFrame(np.random.normal(size=(100, 2), loc=(10, 20), scale=(5, 5)), columns=['x', 'y'])
data2 = pd.DataFrame(np.random.normal(size=(100, 2), loc=(30, 40), scale=(5, 5)), columns=['x', 'y'])
data3 = pd.DataFrame(np.random.normal(size=(100, 2), loc=(32, 17), scale=(5, 5)), columns=['x', 'y'])
```

```python
plt.scatter(data1['x'], data1['y'], c='r', s=10)
plt.scatter(data2['x'], data2['y'], c='g', s=10)
plt.scatter(data3['x'], data3['y'], c='b', s=10);
```

```python
def get_contour(data):
    borders = data.apply(lambda col: [np.min(col), np.max(col)], axis=0)
    return borders
```

```python
def draw_contour1(data, colour):
    borders = get_contour(data)
    print(borders)
    x_min, x_max = borders['x'][0], borders['x'][1]
    y_min, y_max = borders['y'][0], borders['y'][1]
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    plt.pcolormesh(xx, yy, np.zeros(xx.shape), color=colour, alpha=0.01)
```

```python
def draw_contour(data, colour):
    borders = get_contour(data)
    print(borders)
    x_min, x_max = borders['x'][0], borders['x'][1]
    y_min, y_max = borders['y'][0], borders['y'][1]
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    data_full = np.c_[xx.ravel(), yy.ravel()]
    data_full = pd.DataFrame(data=data_full, columns=data.columns)
    plt.scatter(data_full['x'], data_full['y'], c=colour, alpha=0.01, s=10)
```

```python
def check_point_in_contour(row, borders):
    if (borders['x'][0] <= row['x'] <= borders['x'][1]
        and borders['y'][0] <= row['y'] <= borders['y'][1]):
        return True
    return False
```

```python
def check_points_in_other_contours(datas_lst):
    borders_lst = list(map(get_contour, datas_lst))
    
    for i, data in enumerate(datas_lst):
        print("\nData{}:".format(i + 1))
        for j in [k for k in range(len(datas_lst)) if k != i]:
            alien_points = data.apply(check_point_in_contour, axis=1, borders=borders_lst[j])
            alien_points = alien_points[alien_points == True]
            print("In contour of data{} exist points: {}".format(j + 1, alien_points.index.values))
```

```python
plt.scatter(data1['x'], data1['y'], c='r', s=10)
plt.scatter(data2['x'], data2['y'], c='g', s=10)
plt.scatter(data3['x'], data3['y'], c='b', s=10)
draw_contour(data1, 'r')
draw_contour(data2, 'g')
draw_contour(data3, 'b')
```

```python
check_points_in_other_contours([data1, data2, data3])
```
